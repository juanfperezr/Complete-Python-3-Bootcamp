```python
import shutil
import os
import re
```

```python
shutil.unpack_archive('unzip_me_for_instructions.zip','instructions','zip')
```

```python
path = 'C:\\Users\\TG174AC\\source\\repos\\Complete-Python-3-Bootcamp\\12-Advanced Python Modules\\08-Advanced-Python-Module-Exercise\\instructions\\extracted_content'
```

```python
for folder, sub_folders,files in os.walk(path):
    
    print(f"Current folder: {folder}")
    print("\n")
    
    for sub_fold in sub_folders:
        
        print(f"\tCurrent sub_folder: {sub_fold}")
        
    print("\n")    
    for file in files:
        
        print(f"\tCurrent file: {file}")
        
        f = open(folder+'\\'+file,'r')
        text = f.read()
        results = re.findall(r'\d{3}-\d{3}-\d{4}',text)
        if results:
            print(results)
            print('\n')
        else:
            print('No telephone\n')
```
